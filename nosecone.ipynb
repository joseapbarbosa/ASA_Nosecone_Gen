Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```


Geometric Variables

```python
N = 200
L = 160
R = 40
C = 1/3
```


Curve

```python
x = np.linspace(0, 1, N)
x = L * x**2
th = np.arccos(1 - 2 * x / L)
y = R / np.sqrt(np.pi) * np.sqrt(th - np.sin(2 * th) / 2 + C * np.sin(th) ** 3)
C = np.array([x, y])
C = np.transpose(C)
C = np.round(C, 3)
```


File

```python
with open("nosecone.txt", "w") as f:
    f.write("3d=true\nPolyline=true\n")
    for line in C:
        f.write("0.0,{},{}\n".format(line[0], line[1]))
```


Plotting

```python
plt.plot(x, y)
plt.xlim(0, L)
plt.ylim(0, 1.2 * R)
plt.grid()
plt.show()
```
